```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv('/Users/congxiaozhang/Documents/Master Thesis/Thesis/Code/test data with wide stepsize/odd series/zeta_omega_fiveTerms_testdat.csv')
```

```python
df
```

```python
zeta = np.array(df.zeta)
#type(zeta)
OR = np.array(df.omega_ratio)
#type(OR)
power = np.array(df.power)
#type(power)
flag = np.array(df.test_FLAG)
#flag
```

```python
z = zeta.reshape(20,20)
ratio = OR.reshape(20,20)
P = power.reshape(20,20)
f = flag.reshape(20,20)
```

```python
#plot section

plt.xlabel('zeta')
plt.ylabel('Omega Ratio')
plt.title('test data with wide stepsize;odd series;fiveTerms case')

im = plt.contourf(z,ratio,P)
plt.colorbar(im)
for i in range(0,20):
    for j in range(0,20):
        if f[i,j] == 'false':
            plt.plot(z[i,j],ratio[i,j], ".", markersize=5, color = "#483D8B")
        else:
            plt.plot(z[i,j],ratio[i,j], ".", markersize=5, color = "#8B2323")

plt.tight_layout()
plt.savefig('test_plot_OS_fiveTerms')
plt.show()
```
